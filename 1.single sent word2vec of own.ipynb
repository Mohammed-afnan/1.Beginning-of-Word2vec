```python
paragraph = "i love india. i love to play cricket"
```

```python
words = paragraph.lower().split()
print(words)

#for i, word in enumerate(words):
  #print(i,word)
```

```python
import numpy as np
def word2onehot(word_ind,vector):
  #word_vec = np.zeros(vector,dtype=int)
  word_vec = [0 for i in range(vector)]
  word_index = [i for i in range(vector)]
  word_indexi = word_index[word_ind]
  word_vec[word_indexi] = 1
  return word_vec

target_word_veclist = []
for i,word in enumerate(words):
  target_word_vec = word2onehot(i,len(words))
  target_word_veclist.append(target_word_vec)
print(target_word_veclist)
#target_word_veclist = np.asarray(target_word_veclist)
#print(target_word_veclist)
```

```python
dict = {}
for i in range(len(words)):
  target_word = words[i]
  if target_word not in dict:
    context_word = []
    for j in range(i,len(words)):
      if target_word == words[j] and j!=len(words)-1:
        context_word.append(target_word_veclist[j+1])
    dict[target_word] = context_word
print(dict)

max_length = max([len(i) for i in dict.values()])
print(max_length)

extend_with_zero = [0 for i in range(len(words))]
special_char = [1 for i in range(len(words))]
for i in dict.keys():
  if len(dict[i])==0:
    dict[i].append(special_char)
  if len(dict[i])<max_length:
    for j in range(max_length-1):
      if len(dict[i])<max_length:
        dict[i].append(extend_with_zero)
print(dict)

max_length = max([len(i) for i in dict.values()])
print(max_length)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Dense,Input
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
import numpy as np
x = np.array(list(dict.values()))
x = x.reshape(x.shape[0],x.shape[1]*x.shape[2])
print(x.shape)
print(x)
print(len(x))
print(x[0])
print(len(x[0]))
#for i in range(1,len(var),max_length):
  #add = var[i-1]+var[i]
  #x.append(add)
#print(x)
#print(len(x))
#print(len(x[0]))

create_y = [target_word_veclist[words.index(i)] for i in dict.keys() if i in words]
y=[]
for i in create_y:
  var_y = i+extend_with_zero*(max_length-1)
  y.append(var_y)
y = np.array(y)
print(y.shape)
print(y)
print(len(y))
print(len(y[0]))
```

```python
input = Input(shape=(len(x[0]),))
hidden_layer1 = Dense(200,activation='relu')(input)
output = Dense(len(x[0]),activation='softmax')(hidden_layer1)
#whenonly one dense layer with 2 neurons we got loss:11.5 and acc:16.8
                              #12,9,7,4 we got loss:10.7 and acc:16.8

word2vec = Model(input,output)
word2vec.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
word2vec.summary()
```

```python
h = word2vec.fit(x,y,epochs=10,verbose=1)
                 #batch_size=2)
```
